# Credit Card Fraud Data Analysis

## Overview

The usage of card transactions has increased greatly in the last decade with card payments being the leading form of payment as reported by European Central Bank. Criminals are also taking advantage of the new technologies with many adults being affected by card fraud. In the UK alone the payment card fraud makes up 56% of all financial fraud losses in 2018. Data science can be a useful tool for analysing the transactions and their patterns.

(EUB Source: https://www.ecb.europa.eu/press/pr/stats/paysec/html/ecb.pis2018~c758d7e773.en.html)

(UK Card Fraud Source: https://www.ukfinance.org.uk/system/files/Fraud%20The%20Facts%202019%20-%20FINAL%20ONLINE.pdf)

This notebook analyses credit card fraud detection data from Kaggle 

(Link: https://www.kaggle.com/mlg-ulb/creditcardfraud).

The data used in this notebook has been anonymised and transformed using principal component analysis (PCA). In this notebook we will go over the data and find out if we can detect any trends or patterns. Then we will use a number of modelling approaches for classification problems and oversampling methods to figure out which will give us best results according to commonly used metrics.

## Data Modelling 

### Data Preparation 

In the loading step, the data has been divided to 80% train and 20% test with random state of 3 for shuffling the data. Due to the data inbalance the random oversampling algorithm was chosen for the initial evaluation with the models together with scaling.

```python
#Assiging our data and targets, and splitting it into train and test datasets
y = raw_data['Class']
X = raw_data.drop(['Class'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
#Data Oversampling
#RandomOverSampler will be used for this section as it is faster to compute.
over_sampler = RandomOverSampler()
x_res, y_res = over_sampler.fit_resample(x_train, y_train)
```

```python
#Scaling Data
scaler = StandardScaler()
scaler.fit(x_res)
x_scale = scaler.transform(x_res)
x_test_scale = scaler.transform(x_test)
```

### Evaluating Different Models 

In this section we will evaluate different models and their relative performances against selected metrics. For our models apart from the accuracy it is also useful to look into the recall, precision and F1 scores to see how well our model is doing in the overall classification.

```python
#Tensorflow Neural Network Creation Function
def create_model():
    model = Sequential()
    model.add(Dense(128, input_dim=x_scale.shape[1], activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(2, activation='sigmoid'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
#Quick Test of All Models
models = []
names, accuracies, recalls, precisions, f1scores, ROCScores, times = [], [], [], [], [], [], []
models.append(('K Neighbors', KNeighborsClassifier(n_jobs=-1)))
models.append(('Logistic Regression', LogisticRegression(max_iter=10000)))
models.append(('Gaussian NB', GaussianNB()))
models.append(('RandomForest', RandomForestClassifier(n_jobs=-1)))
models.append(('XGB', XGBClassifier(n_jobs=-1)))
models.append(('LGBM', LGBMClassifier(n_jobs=-1)))
models.append(('Neural Network', KerasClassifier(build_fn=create_model, epochs=2, verbose=3)))

for name, model in models:
    start_time = time.time()
    model.fit(x_scale, y_res)
    preds = model.predict(x_test_scale)
    end_time = time.time()
    names.append(name)
    accuracies.append(accuracy_score(y_test, preds))
    recalls.append(recall_score(y_test, preds))
    precisions.append(precision_score(y_test, preds))
    f1scores.append(f1_score(y_test, preds, average='macro'))
    ROCScores.append(roc_auc_score(y_test, preds))
    times.append(end_time - start_time)
```

```python
#Model Results Frame
tests = pd.DataFrame(data={
    'Name':names, 
    'Accuracy':accuracies, 
    'Recall':recalls, 
    'Precision':precisions, 
    'F1 Score (Macro)':f1scores, 
    'ROC Score':ROCScores, 
    'Time (s)':times
})
tests
```

The model with the highest recall and ROC value is logistic regression with smallest amount of false negatives. However its precision is very poor with a lot of false positives which is shown in the F1 score.

RandomForest, XGB and LGBM perform best in terms of accuracy, precision, ROC and F1 scores. XGB performs best in the F1 score where XGB has a marginally better F1 score and LGBM has a marginally better recall.

The neural network appears to be the best in between solution where it has a higher recall and relatively high F1 score.

Based on the results it will likely be worth investigating XGB, LGBM and Neural Network for further refinement. Also we can try to look at the logistic regression parameters to see if it can be improved further.

### Optimising Models 

In this section we will iterate over different paramets of the models to find out which is performing best and run them. The voting classifier to test if better performance can be obtained.

In interest of saving time we will optimize the parameters one at a time instead of using GridSearchCV (XGB Classifier can easily take over 2 hours when we iterate over all the combinations!). This way we can also see how changes in the parameters affect model predictions.

```python
#Logistic Regression Model Optimisation
penalty_params = ['l2', 'none'] #l1, 'elasticnet'
c_params = [0.002, 0.02, 0.2, 0.5, 1.0, 2.0, 5.0]
solver_params = ['newton-cg', 'lbfgs', 'liblinear', 'newton-cg','sag','saga']
multi_class_params = ['auto', 'ovr', 'multinomial']
all_params = [penalty_params, c_params, solver_params, multi_class_params]
param_switch = ['penalty', 'c', 'solver', 'multi-class']

par, ch, accuracies1, recalls1, precisions1, f1scores1, ROCScores1 = [], [], [], [], [], [], []

def param_checker(params, param_sw):
    for param in params:
        if param_sw == 'penalty':
            model = LogisticRegression(max_iter=1000, penalty=param)
        elif param_sw == 'c':
            model = LogisticRegression(max_iter=1000, C=param)
        elif param_sw == 'solver':
            model = LogisticRegression(max_iter=1000, solver=param)
        elif param_sw == 'multi-class':
            model = LogisticRegression(max_iter=1000, multi_class=param)

        kfold = KFold(n_splits=5, random_state=1)
        imba_pipeline = make_pipeline(RandomOverSampler(), StandardScaler(), model)
        cv_dict = cross_validate(imba_pipeline, x_train, y_train,
                         scoring=['accuracy', 'recall', 'precision', 'f1_macro', 'roc_auc'], cv=kfold, n_jobs=-1)

        par.append(param_sw)
        ch.append(param)
        accuracies1.append(sum(cv_dict['test_accuracy'])/len(cv_dict['test_accuracy']))
        recalls1.append(sum(cv_dict['test_recall'])/len(cv_dict['test_recall']))
        precisions1.append(sum(cv_dict['test_precision'])/len(cv_dict['test_precision']))
        f1scores1.append(sum(cv_dict['test_f1_macro'])/len(cv_dict['test_f1_macro']))
        ROCScores1.append(sum(cv_dict['test_roc_auc'])/len(cv_dict['test_roc_auc']))
    return None

for i in range(4):
    param_checker(all_params[i], param_switch[i])
```

```python
#Logistic Regression Results Dataframe
logregression_frame = pd.DataFrame({
    'Parameter Type':par,
    'Parameter Value':ch,
    'Accuracy':accuracies1,
    'Recall':recalls1,
    'Precision':precisions1,
    'F1 Score (Macro)':f1scores1,
    'ROC Score':ROCScores1,
})
logregression_frame.sort_values(by=['Parameter Type', 'Parameter Value'])
```

```python
#Optimal Logistic Regression Parameters
start_time = time.time()
model = LogisticRegression(max_iter=1000, C=0.002)
model.fit(x_scale, y_res)
preds = model.predict(x_test_scale)
print('Confusion Matrix:')
print(confusion_matrix(y_test, preds))
print('Recall', recall_score(y_test, preds))
print('Precision', precision_score(y_test, preds))
print('F1 Score', f1_score(y_test, preds, average='macro'))
print('ROC Score', roc_auc_score(y_test, preds))
end_time = time.time()
print('Time Taken: {:.2f}s'.format(end_time-start_time))
```

The parameter C has a significant impact on the model precision with 0.002 producing the best results. The rest of the parameters can be kept at the default to achieve those results.

```python
#XGB Model Optimisation
treemethod = ['exact', 'approx', 'hist']
maxdepth = [i for i in range(6, 10, 1)]
gammas = [i for i in range(1, 11, 2)]
param_sw = ['treemethod', 'maxdepth', 'gammas']
all_params = [treemethod, maxdepth, gammas]
par2, ch2, accuracies2, recalls2, precisions2, f1scores2, ROCScores2 = [], [], [], [], [], [], []

def param_checker(params, param_sw):
    for param in params:
        print(param, param_sw) #XGB takes a long time to run, this command serves as a step tracker. 
        if param_sw == 'treemethod':
            model = XGBClassifier(tree_method=param, n_jobs=-1)
        elif param_sw == 'maxdepth':
            model = XGBClassifier(max_depth=param, n_jobs=-1)
        elif param_sw == 'gammas':
            model = XGBClassifier(gamma=param, n_jobs=-1)
            
        kfold = KFold(n_splits=5, random_state=1)
        imba_pipeline = make_pipeline(RandomOverSampler(), StandardScaler(), model)
        cv_dict = cross_validate(imba_pipeline, x_train, y_train,
                         scoring=['accuracy', 'recall', 'precision', 'f1_macro', 'roc_auc'], cv=kfold, n_jobs=-1)
        
        par2.append(param_sw)
        ch2.append(param)
        accuracies2.append(sum(cv_dict['test_accuracy'])/len(cv_dict['test_accuracy']))
        recalls2.append(sum(cv_dict['test_recall'])/len(cv_dict['test_recall']))
        precisions2.append(sum(cv_dict['test_precision'])/len(cv_dict['test_precision']))
        f1scores2.append(sum(cv_dict['test_f1_macro'])/len(cv_dict['test_f1_macro']))
        ROCScores2.append(sum(cv_dict['test_roc_auc'])/len(cv_dict['test_roc_auc']))
    return None

for i in range(len(param_sw)):
    param_checker(all_params[i], param_sw[i])
```

```python
#XGB Results Dataframe
xgb_frame = pd.DataFrame({
    'Parameter':par2,
    'Value':ch2,
    'Accuracy':accuracies2,
    'Recall':recalls2,
    'Precision':precisions2,
    'F1 Score (Macro)':f1scores2,
    'ROC Score':ROCScores2,
})
xgb_frame.sort_values(by=['Parameter', 'Value'])
```

```python
#LGBM Model Optimisation
boosttype = ['gbdt', 'dart', 'goss']
lrs = [0.2, 0.1, 0.05, 0.02, 0.01, 0.001]
estims = [i for i in range(1, 1001, 100)]
all_params = [boosttype, lrs]
param_sw = ['boosttype', 'lrs']

par3, ch3, accuracies3, recalls3, precisions3, f1scores3, ROCScores3 = [], [], [], [], [], [], [],

def param_checker(params, param_sw):
    for param in params:
        if param_sw == 'boosttype':
            model = LGBMClassifier(boosting_type=param, n_jobs=-1)
        elif param_sw == 'lrs':
            model = LGBMClassifier(learning_rate=param, n_jobs=-1)
    
        kfold = KFold(n_splits=5, random_state=1)
        imba_pipeline = make_pipeline(RandomOverSampler(), StandardScaler(), model)
        cv_dict = cross_validate(imba_pipeline, x_train, y_train,
                         scoring=['accuracy', 'recall', 'precision', 'f1_macro', 'roc_auc'], cv=kfold, n_jobs=-1)
        
        par3.append(param_sw)
        ch3.append(param)
        accuracies3.append(sum(cv_dict['test_accuracy'])/len(cv_dict['test_accuracy']))
        recalls3.append(sum(cv_dict['test_recall'])/len(cv_dict['test_recall']))
        precisions3.append(sum(cv_dict['test_precision'])/len(cv_dict['test_precision']))
        f1scores3.append(sum(cv_dict['test_f1_macro'])/len(cv_dict['test_f1_macro']))
        ROCScores3.append(sum(cv_dict['test_roc_auc'])/len(cv_dict['test_roc_auc']))
    return None

for i in range(len(param_sw)):
    param_checker(all_params[i], param_sw[i])
```

```python
#Keras Model Optimisation
neurons = [16, 32, 64, 128, 256, 512]
activations = ['relu', 'selu', 'softmax','sigmoid']
all_params = [neurons, activations]

ns, acts, accuracies4, recalls4, precisions4, f1scores4, ROCScores4 = [], [], [], [], [], [], [] 

def keras_iter(a, b):
    def create_model():
        model = Sequential()
        model.add(Dense(a, input_dim=x_scale.shape[1], activation=b))
        model.add(Dense(a, activation=b))
        model.add(Dense(a, activation=b))
        model.add(Dense(2, activation='softmax'))
        model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
        return model

    estimator = KerasClassifier(build_fn=create_model, epochs=2, verbose=2)
    estimator.fit(x_scale, y_res)
    preds = estimator.predict(x_test_scale)
    ns.append(a)
    acts.append(b)
    accuracies4.append(accuracy_score(y_test, preds))
    recalls4.append(recall_score(y_test, preds))
    precisions4.append(precision_score(y_test, preds))
    f1scores4.append(f1_score(y_test, preds, average='macro'))
    ROCScores4.append(roc_auc_score(y_test, preds))
```

```python
for n in neurons:
    keras_iter(n, 'relu')

for a in activations:
    keras_iter(64, a)
```

```python
#Neural Network Results Dataframe
nn_frame = pd.DataFrame({
    'Neurons':ns,
    'Activation':acts,
    'Accuracy':accuracies4,
    'Recall':recalls4,
    'Precision':precisions4,
    'F1 Score (Macro)':f1scores4,
    'ROC Score':ROCScores4,
})
nn_frame
```

```python
#Optimal Keras Model
start_time = time.time()
def create_model():
    model = Sequential()
    model.add(Dense(512, input_dim=x_scale.shape[1], activation='relu'))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(2, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

estimator = KerasClassifier(build_fn=create_model, epochs=2, verbose=2)
estimator.fit(x_scale, y_res)
preds = estimator.predict(x_test_scale)
print('Confusion Matrix')
print(confusion_matrix(y_test, preds))
print('Recall', recall_score(y_test, preds))
print('Precision', precision_score(y_test, preds))
print('F1 Score', f1_score(y_test, preds, average='macro'))
print('ROC Score', roc_auc_score(y_test, preds))
end_time = time.time()
print('Time Taken: {:.2f}s'.format(end_time-start_time))
```

The results from the neural network are satisfactory, but they do not perform as well as the other models in terms of best recall or the best accuracy and F1 score. For this reason we will use other models in further evaluations. In similarity to the rest of the algorithms as we change the parameters of the neural network the bigger the trade-off between the recall and ROC, and the rest of the parameters. 

```python
#Voting Classifier
start_time = time.time()
models = []
models.append(('LRC', LogisticRegression(max_iter=1000, C=0.002)))
models.append(('XGB', XGBClassifier(n_estimators=100, n_jobs=-1, max_depth=7)))
models.append(('LGB', LGBMClassifier(n_estimators=100, n_jobs=-1, learning_rate=0.2)))
elf = VotingClassifier(estimators=models, voting='soft', weights=(1, 1, 1), n_jobs=-1)
elf.fit(x_scale, y_res)
preds = elf.predict(x_test_scale)
print('Confusion Matrix: ')
print(confusion_matrix(y_test, preds))
print('Recall', recall_score(y_test, preds))
print('Precision', precision_score(y_test, preds))
print('F1 Score', f1_score(y_test, preds, average='macro'))
print('ROC Score', roc_auc_score(y_test, preds))
end_time = time.time()
print('Time Taken: {:.2f}s'.format(end_time-start_time))
```

By using the voting classifier it is possible combine the predictive powers of the combined models by using weighting. In case of the even weighted soft vote the voting classifier performs at a level similar to LGBM and XGB models.

### Evaluation of Different Sampling Techniques

To improve the model performance further we can test different oversampling techniques if they produce better results with different models. In this section we will iterate over the different oversampler and model combinations.

```python
#Models for Oversampling Test
over_models = []
over_models.append(('XGB', XGBClassifier(n_estimators=100, n_jobs=-1, max_depth=7)))
over_models.append(('LGB', LGBMClassifier(n_estimators=100, n_jobs=-1, learning_rate=0.2)))
over_models.append(('LGR', LogisticRegression(max_iter=1000)))
```

```python
#List of Oversamplers
over_list = []
over_list.append(('ADASYN', ADASYN(n_jobs=-1)))
over_list.append(('Borderline SMOTE', BorderlineSMOTE(n_jobs=-1)))
over_list.append(('SMOTE', SMOTE(n_jobs=-1)))
over_list.append(('Random Oversampling', RandomOverSampler()))
```

```python
#For Loop for Testing Oversamplers against the Models
names_m, names_o, accuracies, recalls, precisions, f1scores, ROCScores, times = [], [], [], [], [], [], [], []
scaler = StandardScaler()
for name_o, sampler in over_list:
    print('Running: ', name_o)
    for name_m, model in over_models:
        kfold = KFold(n_splits=5, random_state=1)
        imba_pipeline = make_pipeline(sampler, StandardScaler(), model)
        cv_dict = cross_validate(imba_pipeline, x_train, y_train,
                         scoring=['accuracy', 'recall', 'precision', 'f1_macro', 'roc_auc'], cv=kfold, n_jobs=-1)
        names_m.append(name_m)
        names_o.append(name_o)
        accuracies.append(sum(cv_dict['test_accuracy'])/len(cv_dict['test_accuracy']))
        recalls.append(sum(cv_dict['test_recall'])/len(cv_dict['test_recall']))
        precisions.append(sum(cv_dict['test_precision'])/len(cv_dict['test_precision']))
        f1scores.append(sum(cv_dict['test_accuracy'])/len(cv_dict['test_accuracy']))
        ROCScores.append(sum(cv_dict['test_roc_auc'])/len(cv_dict['test_roc_auc']))
```

```python
#Oversampling Results Dataframe
over_sample_frame = pd.DataFrame({
    'Model Name':names_m,
    'Oversampler Name':names_o,
    'Accuracy':accuracies,
    'Recall':recalls,
    'Precision':precisions,
    'F1 Score (Macro)':f1scores,
    'ROC Score':ROCScores,
})
over_sample_frame.sort_values(by='Model Name')
```

The tests results show mixed results for our metrics. There is a clear inverse relationship between the recall and ROC scores when compared to accuracy, precision and F1 scores. It seems that as the models are more optimized for detecting fraudulent transactions they start to produce a larger number of false positives and vice versa.

The best model for recall and ROC score is Logistic Regression coupled with random oversampling or ADASYN.

The best model for general accuracy and F1 Score is XGB classifier coupled with boderline SMOTE or random oversampling.

The selection of which model should be used depends on what is most important for the customer. While logistic regression model can greatly reduce the amount of fraudulent transactions it will also stop transactions of large number of customers, while the XGB model would allow for better overall classification accuracy.
